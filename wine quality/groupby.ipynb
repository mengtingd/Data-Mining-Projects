```python
import pandas as pd

wine_df=pd.read_csv('winequality_edited.csv')
wine_df=wine_df.iloc[:,1:]
wine_df.head()
```

```python
wine_df.mean()
```

```python
wine_df.groupby('quality').mean()
```

```python
wine_df.groupby(['quality','color']).mean()
```

```python
wine_df.groupby(['quality','color'],as_index=False)['pH'].mean()
```

```python
#按照颜色分组，找到每组的平均质量等级
#某种类型的葡萄酒是否代表更高的品质？
wine_df.groupby('color',as_index=False)['quality'].mean()
```

```python
#描述功能，查看最小、25%、50%、75% 和 最大 pH 值
wine_df.describe()
```

```python
# 对用于把数据“分割”成组的边缘进行分组
bin_edges = [2.72,3.11,3.21,3.32,4.01]

# 四个酸度水平组的标签
bin_names = ['高','中等偏高','中','低']

# 创建 acidity_levels 列
wine_df['acidity_levels'] = pd.cut(wine_df['pH'], bin_edges, labels=bin_names)

# 检查该列是否成功创建
wine_df.head()
```

```python
# 用 groupby 计算每个酸度水平的平均质量
wine_df.groupby('acidity_levels',as_index=False)['quality'].mean()
```

```python
# 保存更改，供下一段使用
wine_df.to_csv('winequality_edited2.csv', index=False)
```
